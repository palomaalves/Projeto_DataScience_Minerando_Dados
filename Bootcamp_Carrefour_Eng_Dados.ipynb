# Governança de dados

Quando falamos em modelos de área de governança de dados, quais são os mais conhecidos?

Os modelos de governança podem varias de um modelo mais centralizado, passando por modelos híbridos, modelos mais descentralizados até modelos distribuídos.

O que é uma governança centralizada?

O modelo de governança centralizada prevê que existam padrões e políticas corporativas, mas que compete a uma área central garantir sua aplicação em todo o ecossistema

O que é Governança?

A governança é a organização sobre todos os processos de um sistema social, seja através de leis, normas, poder ou linguagem.

O que é Governança de dados?

A governança é a organização sobre todos os processos de um ecossistema de dados, seja através de leis, normas, poder ou linguagem.

Se os dados fossem um carro, que parte governança seria?

Freios - é o conjunto de peças que garante que se possa acelerar com a segurança e sem acidentes.

O que é uma governança híbrida?

O modelo de governança híbrida prevê que existam padrões e políticas corporativas, mas que sua aplicação e monitoramento sejam compartilhados entre todos os participantes do ecossistema.

O que é uma governança descentralizada?

O modelo de governança descentralizada prevê que existam padrões e políticas corporativas, mas que compete a cada participante do ecossistema garantir sua aplicação.

Nos “P” patrocínio estamos falando de?

Apoio executivo em todas as instancias

Quais os “Ps” para se planejar uma área de governança?

PROGRAMA, PRINCÍPIOS, PROPÓSITO, PATROCÍNIO, PLANEJAMENTO, POSICIONAMENTO, PRODUTO, PROTÓTIPO, PILOTO, PROJETO.

Quais os “Ps” para se viabilizar a cultura de entrega de produtos de valor para os clientes?

PROTÓTIPO, PILOTO, PROJETO.

# Teoria e Prática com YARN

Sobre YARN podemos dizer que:
Faz o gerenciamento de recursos e monitoramento de Jobs, onde os recursos dos nós são alocados somente quando requisitados (via container).

Qual é a diferença entre um nó normal e um nó master?

Um nó é computador individual no cluster, já o nó máster é um driver que gerencia a distribuição para os workers.

Como diferenciamos os conceitos de escalabilidade vertical e horizontal?

Escalabilidade vertical é o aumento de processamento tradicional, onde a capacidade do processador, memória e throughput de rede são aumentados para melhorar a potência computacional da máquina. Já na escalabilidade horizontal, ao invés de aumentarmos a capacidade de apenas uma máquina, acoplamos máquinas menores em rede para aumentar a capacidade de processamento de forma distribuída.

O que é um cluster?

Um grupo de computadores que trabalham juntos, provendo armazenamento, processamento e gerenciamento de recursos.


Como fazemos a cópia de arquivos no HDFS?

Exemplo: $ hdfs dfs –get /tmp/file_teste.text

O que é um daemon?

Um programa (serviço) rodando nos nós (normal e master). Cada daemon possui uma função específica no cluster.

Podemos definir o Hadoop como:

Projeto de software open-source escalável capaz de lidar com falhas de hardware, software e rede, além de processamento distribuído. Sendo assim pode ser utilizado como um framework para computação distribuída, também conhecido como "SO de Big Data".

O que é Big Data?

Big Data é um processo de análise e interpretação de um grande volume de dados armazenados remotamente, podendo consolidar dados coletados nos mais diversos domínios de aplicação.

Sobre HDFS, podemos afirmar que:

É baseado no Google FS; projeto escalável e tolerante a falhas, trata de arquivos texto, sequence file, Parquet, AVRO, ORC, etc. Possui um tamanho mínimo de bloco de 128mb e fator de replicação (default:3).

Sobre os componentes do HDFS, temos:

NameNode (gerencia o namespace), DataNode (armazena blocos de arquivos), Secondary NameNode (oferece tarefas de ponto de verificação e manutenção do namenode).

# Pacotes necessários

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

# Lendo o banco de dados

```python
dados=pd.read_csv('/content/train.csv')
```

```python
dados.head()
```

```python
dados = dados.drop(['Name', 'Ticket', 'Cabin', 'Embarked'], axis=1)
dados.head()
```

#Editando chave e variaável resposta

```python
dados = dados.set_index(['PassengerId'])
dados = dados.rename(columns = {' target': 'target'}, inplace = False)
```

```python
dados.head()
```

#Descritiva

```python
dados.describe()
```

```python
dados.describe(include=['O'])
```

# Transformação de dados

```python
dados['Sex_F']=np.where(dados['Sex']=='female', 1, 0)
dados['Pclass_1']=np.where(dados['Pclass']==1, 1, 0)
dados['Pclass_2']=np.where(dados['Pclass']==2, 1, 0)
dados['Pclass_3']=np.where(dados['Pclass']==3, 1, 0)
```

```python
dados = dados.drop(['Pclass', 'Sex'], axis = 1)
```

```python
dados.head()
```

```python
dados.isnull().sum()
```

```python
dados.fillna(0, inplace = True)
```

# Amostragem

```python
x_train, x_test, y_train, y_test = train_test_split(dados.drop(['target'], axis = 1), 
                                                    dados['target'],
                                                    test_size = 0.3,
                                                    random_state = 1234)
```

```python
[{'treino': x_train.shape}, {'teste': x_test.shape}]
```

# Modelo

```python
rndforest = RandomForestClassifier(n_estimators = 1000,
                                   criterion = 'gini',
                                   max_depth = 5)
rndforest.fit(x_train, y_train)
```

```python
probabilidade = rndforest.predict_proba(dados.drop('target', axis = 1))[:,1]
classificacao = rndforest.predict(dados.drop('target', axis=1))
```

```python
dados['probabilidade']=probabilidade
dados['classificacao']=classificacao
```
